```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as mpl
df = pd.read_csv('C:/Users/lucan/Documents/Uni/Courses/Bachelor Semester Project/BSP1/code/yagoFactsCleaned.csv')
df.columns = ['Subject', 'Predicate', 'Object']
```

Ideally the search function should be able to find paths that would exist in a simple graph representing the yago dataset. Using NetworkX might make it easier to find paths between edges.

```python
def search(queryList):
    df2 = df.loc[df['Subject'].isin(queryList) | df['Object'].isin(queryList)]
    predicateList = df2['Predicate'].tolist()
    predicateListUnique = list(dict.fromkeys(predicateList))
    Graphtype = nx.MultiGraph()
    G = nx.from_pandas_edgelist(df2, 'Subject', 'Object', 'Predicate', Graphtype)
    mpl.figure(figsize=(16, 8))
    nx.draw_networkx(G, with_labels=True)
```

```python
search(['Ricochet_(TV_production_company)', 'Sahara'])
```
